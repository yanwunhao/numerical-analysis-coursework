```python
def fv(v):
    return 2*v - 0.16814 + 12.2/v - 1.0102314/(v*v) - 30.77025
```

```python
sampling_v = []
sampling_Fv = []
for i in range(10, 200, 1):
    input_v = i * 0.1
    sampling_v.append(input_v)
    output = fv(input_v)
    sampling_Fv.append(output)

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(sampling_v, sampling_Fv)

ax.grid()

plt.show()
```

```python
def binarysearch(left_bound, right_bound):
    left = left_bound
    right = right_bound
    for i in range(10):
        mid = (left + right)/2.
        fv_left = fv(left)
        fv_mid = fv(mid)
        fv_right = fv(right)
        print(f"Epoch {i}: left point is {left}, right point is {right}")
        if fv_left * fv_mid < 0:
            print(f"Update right bound: {right} -> {mid}")
            right = mid
        else:
            print(f"Update left bound: {left} -> {mid}")
            left = mid
    return (left + right) / 2

result = binarysearch(12, 16)
print(f"The result of binary search is {result}")
```
